```python
from PIL import Image
import numpy as np

def conversion(img):
    return img.convert("RGB")

imagenes=[]

for i in range(0,968):
    if(i<10):
        pri="000"+str(i)
    else:
        if(i<100):
            pri="00"+str(i)
        else:
            pri="0"+str(i)
    ruta ="natural_images\\car\\car_"+pri+".jpg"
    imagenes.append(Image.open(ruta))
    
imagenes=[conversion(img) for img in imagenes]
```

```python
tnsor=np.zeros((968,100,100,3))
for i in range(0,968):
    alto,ancho=imagenes[i].size
    for y in range(alto):
        for x in range(ancho):
            r, g, b = imagenes[i].getpixel((x, y))
            tnsor[i,x,y,0]=r
            tnsor[i,x,y,1]=g
            tnsor[i,x,y,2]=b
```

```python
def distancia(ten1,ten2):
   aux=abs(ten1-ten2)
   suma=0
   count=0
   for i in range(len(aux)):
      for j in range(len(aux[i])):
         for k in range(len(aux[i][j])):
            suma=suma+aux[i][j][k]
            count=count+1
   return suma/count
```

```python
k1=20*np.ones((100,100,3))
k2=80*np.ones((100,100,3))
k3=100*np.ones((100,100,3))
k4=2*np.ones((100,100,3))
k5=3*np.ones((100,100,3))
k6=7*np.ones((100,100,3))



for epoca in range(300):
    clas1=[]
    clas2=[]
    clas3=[]
    clas4=[]
    clas5=[]
    clas6=[]

    for t in tnsor:

        d1=distancia(t,k1)
        d2=distancia(t,k2)
        d3=distancia(t,k3)
        d4=distancia(t,k4)
        d5=distancia(t,k5)
        d6=distancia(t,k6)
        
        
        if(d1<d2 and d1<d3 and d1<d4 and d1<d5 and d1<d6):
            clas1.append(t)
        if(d2<d1 and d2<d3 and d2<d4 and d2<d5 and d2<d6):
            clas2.append(t)
        if(d3<d1 and d3<d2 and d3<d4 and d3<d5 and d3<d6):
            clas3.append(t)
        if(d4<d2 and d4<d3 and d4<d1 and d4<d5 and d4<d6):
            clas4.append(t)
        if(d5<d1 and d5<d3 and d5<d4 and d5<d2 and d5<d6):
            clas5.append(t)
        if(d6<d1 and d6<d2 and d6<d4 and d6<d5 and d6<d3):
            clas6.append(t)


    k1_=k1
    k2_=k2
    k3_=k3
    k4_=k4
    k5_=k5
    k6_=k6

    
    #Promediar los elementos dentro de cada cluster
    if len(clas1)>0 : 
        aux=np.zeros((100,100,3))
        count=0
        for i in clas1:
            aux=aux+i
            count=count+1
        k1_=aux/count
    if len(clas2)>0:
        aux=np.zeros((100,100,3))
        count=0
        for i in clas2:
            aux=aux+i
            count=count+1
        k2_=aux/count
    if len(clas3)>0:
        aux=np.zeros((100,100,3))
        count=0
        for i in clas3:
            aux=aux+i
            count=count+1
        k3_=aux/count

    if len(clas4)>0 : 
        aux=np.zeros((100,100,3))
        count=0
        for i in clas4:
            aux=aux+i
            count=count+1
        k4_=aux/count
    if len(clas5)>0:
        aux=np.zeros((100,100,3))
        count=0
        for i in clas5:
            aux=aux+i
            count=count+1
        k5_=aux/count
    if len(clas6)>0:
        aux=np.zeros((100,100,3))
        count=0
        for i in clas6:
            aux=aux+i
            count=count+1
        k6_=aux/count
    


    if(distancia(k1,k1_)<0.001 and distancia(k2,k2_)<0.001 and distancia(k3,k3_)<0.001 and distancia(k4,k4_)<0.001 and distancia(k5,k5_)<0.001 and distancia(k6,k6_)<0.001):
        break
    else:
        k1=k1_
        k2=k2_
        k3=k3_
        k4=k4_
        k5=k5_
        k6=k6_



```

```python
from math import trunc
efe=k1
for i in range(len(efe)):
    for j in range(len(efe[i])):
        for r in range(len(efe[i][j])):
            efe[i][j][r]=trunc(efe[i][j][r])
            if efe[i][j][r]>255:
                efe[i][j][r]=255
imagen_centroide1 = Image.fromarray(efe.astype(np.uint8)) 

efe=k2
for i in range(len(efe)):
    for j in range(len(efe[i])):
        for r in range(len(efe[i][j])):
            efe[i][j][r]=trunc(efe[i][j][r])
            if efe[i][j][r]>255:
                efe[i][j][r]=255
imagen_centroide2 = Image.fromarray(efe.astype(np.uint8)) 
efe=k3
for i in range(len(efe)):
    for j in range(len(efe[i])):
        for r in range(len(efe[i][j])):
            efe[i][j][r]=trunc(efe[i][j][r])
            if efe[i][j][r]>255:
                efe[i][j][r]=255
imagen_centroide3 = Image.fromarray(efe.astype(np.uint8)) 


efe=k4
for i in range(len(efe)):
    for j in range(len(efe[i])):
        for r in range(len(efe[i][j])):
            efe[i][j][r]=trunc(efe[i][j][r])
            if efe[i][j][r]>255:
                efe[i][j][r]=255
imagen_centroide4 = Image.fromarray(efe.astype(np.uint8)) 

efe=k5
for i in range(len(efe)):
    for j in range(len(efe[i])):
        for r in range(len(efe[i][j])):
            efe[i][j][r]=trunc(efe[i][j][r])
            if efe[i][j][r]>255:
                efe[i][j][r]=255
imagen_centroide5 = Image.fromarray(efe.astype(np.uint8)) 
efe=k6
for i in range(len(efe)):
    for j in range(len(efe[i])):
        for r in range(len(efe[i][j])):
            efe[i][j][r]=trunc(efe[i][j][r])
            if efe[i][j][r]>255:
                efe[i][j][r]=255
imagen_centroide6 = Image.fromarray(efe.astype(np.uint8)) 
```
